```python
import tkinter as tk
import random
from random import randrange
import numpy as np
from PIL import ImageTk , Image


fen=tk.Tk()
nb_l=4 
nb_c=5
dim=125
nombres=['zero','un','deux','trois','quatre','cinq','six','sept','huit','neuf']  
global cpt
global elem
global im 
global cartes 
global affich
im=tk.PhotoImage(file='image/carre.gif')


    
def melanger_grille():
    cartes= list(range(nb_l*nb_c//2))*2
    random.shuffle(cartes)
    cartes= np.reshape(cartes,(nb_l,nb_c))
    return cartes 
    

def list_photo():
    affich=[]
    for nb in nombres:
        fichier='image/' +nb+ '.gif'
        img = ImageTk.PhotoImage(Image.open(fichier))  # PIL solution
        affich.append(img)
    return affich


can= tk.Canvas(fen,width=dim*nb_c,heigh=dim*nb_l,bg='gray')
can.pack(side=tk.LEFT)
btn = tk.Button(fen , text='Nouveau')

btn.pack(padx=10,pady=10)
lbl=tk.Label(fen,text=0,font='courier 20 bold')
lbl.pack(padx=20,pady=20)

  #afficher la grille initiale
def init():
    global elem
    global cartes
    global affich
    elem=[None , None]
    global cpt ,score
    cpt=0
    global im
    cartes=melanger_grille()
    affich= list_photo()
    for i in range(nb_c):
        for j in range(nb_l):
            centre=(i*dim+dim/2,j*dim+dim/2)
            can.create_image(centre,image=im)
    lbl['text']=0
    score=0
    return cartes,affich

cartes,affich=init()
#---------------------------
        
centre=(randrange(nb_l*dim),randrange(nb_l*dim))


def clic(event):
    global cpt
    global elem
    global score

    if(elem[1]!=None): 
        return
    x=event.x
    y=event.y
    j=x//dim
    i=y//dim
    
    if(elem[0]!=None): #cliquer deux fois sur la même case
        if(elem[0]==[i,j]):
            return
        
    cord=[i,j]   
    if(cartes[i,j]>-1):
        cpt=cpt+1 
        elem[cpt-1]=[i,j]
        
        centre=(j*dim+dim/2,i*dim+dim/2)
        can.create_image(centre,image=affich[cartes[i,j]]) #retourner la carte 
        if (cpt==2): #si on clique sur deux éléments 
            verif()
            score=score+1
            lbl['text']=score
        
        
      
    
def verif():
    global elem
    global cpt
    if(cartes[elem[0][0],elem[0][1]]!=cartes[elem[1][0],elem[1][1]]):
        can.after(400,tourner,elem[0][0],elem[0][1])
        can.after(400,tourner,elem[1][0],elem[1][1])


    else: 
        cartes[elem[0][0],elem[0][1]]=-1 
        cartes[elem[1][0],elem[1][1]]=-1
    elem=[None,None]
    cpt=0  
       

 
    
def tourner(i,j):
    global elem
    global im
    centre=(j*dim+dim/2,i*dim+dim/2)
    can.create_image(centre,image=im)

def new(event):
    global cartes
    global affich
    cartes,affich=init()
    
can.bind("<Button>",clic)
btn.bind("<Button>",new)

fen.mainloop()
```
